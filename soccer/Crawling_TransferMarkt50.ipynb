## 한페이지 크롤링

```python
# 라이브러리 import
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# 구글 검색창에 입력한 my users agent 정보를 headers에 설정
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Whale/3.22.205.26 Safari/537.36'}

# 크롤링할 페이지가 있는 주소값을 넣기
url = "https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop"

r = requests.get(url, headers=headers)
r.status_code


# 분석할 html 소스 가져오기
soup = BeautifulSoup(r.text, 'html.parser')
print(soup)

# 선수들의 정보가 담긴 태그와 class_ 찾기
player_info = soup.find_all('tr', class_={'odd', 'even'})

# 첫번째 요소 확인하기
print(player_info[0])

# 전체 개수 확인하기
print("")
print('전체 선수들의 개수: ', len(player_info),'개')

# 가져올 7개의 정보가 담길 빈 리스트 만들기
# number, name, position, age, nation, team, value
number = [] 
name = []
position = []
age = []
nation = []
team = []
value = []

# player_info에서 'td' 태그만 모두 찾기
for info in player_info:
    player = info.find_all("td")
#     print(player)
#     print(player[0])
    
# 해당정보를 찾아서 각 리스트에 .append로 추가하기
    number.append(player[0].text)
    name.append(player[3].text)
    position.append(player[4].text)
    age.append(player[5].text)
    nation.append(player[6].img['alt'])
    team.append(player[7].img['alt'])
    value.append(player[8].text.strip()) #strip() 앞뒤 공백 제거
    

#pd.DataFrame()으로 저장하기

df = pd.DataFrame(
        {"number":number,
         "name":name,
         "position":position,
         "age":age,
         "nation":nation,
         "team":team,
         "value":value}
    )
df.head()

#df.csv() 파일로 저장하기
df.to_csv('transfermark25.csv', index=False)
```

```python
# 구글 검색창에 입력한 my users agent 정보를 headers에 설정
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Whale/3.22.205.26 Safari/537.36'}

number = [] 
name = []
position = []
age = []
nation = []
team = []
value = []
    
for i in range(1, 3):
    # 크롤링할 페이지가 있는 주소값을 넣기
    url = f"https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop?page={i}"

    r = requests.get(url, headers=headers)
    r.status_code

    # 분석할 html 소스 가져오기
    soup = BeautifulSoup(r.text, 'html.parser')

    # 선수들의 정보가 담긴 태그와 class_ 찾기
    player_info = soup.find_all('tr', class_={'odd', 'even'})


    # player_info에서 'td' 태그만 모두 찾기
    for info in player_info:
        player = info.find_all("td")

    # 해당정보를 찾아서 각 리스트에 .append로 추가하기
        number.append(player[0].text)
        name.append(player[3].text)
        position.append(player[4].text)
        age.append(player[5].text)
        nation.append(player[6].img['alt'])
        team.append(player[7].img['alt'])
        value.append(player[8].text.strip()) #strip() 앞뒤 공백 제거

    time.sleep(1)    
    
#pd.DataFrame()으로 저장하기

df = pd.DataFrame(
        {"number":number,
         "name":name,
         "position":position,
         "age":age,
         "nation":nation,
         "team":team,
         "value":value}
    )
df.tail()

#df.csv() 파일로 저장하기
df.to_csv('transfermark50.csv', index=False)
```
